```python
import pandas as pd
import matplotlib.pyplot as plt

Mil = pd.read_csv('execution_1000.csv')
CMil = pd.read_csv('execution_5000.csv')
DMil = pd.read_csv('execution_10000.csv')
QMil = pd.read_csv('execution_15000.csv')
VMil = pd.read_csv('execution_20000.csv')

# Create a figure and axes for the plots
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, figsize=(7, 7))

# Mil insert
ax1.plot(Mil['Ejecucion'], Mil['tiempo_segundos'])
ax1.set_xlabel('Nmero de intentos')
ax1.set_ylabel('Tiempos (s)')
ax1.set_title('Hashing Insert 1000')

# Cinco Mil insert
ax2.plot(CMil['Ejecucion'], CMil['tiempo_segundos'])
ax2.set_xlabel('Nmero de intentos')
ax2.set_ylabel('Tiempos (s)')
ax2.set_title('Hashing Insert 5000')

# Diez Mil insert
ax3.plot(DMil['Ejecucion'], DMil['tiempo_segundos'])
ax3.set_xlabel('Nmero de intentos')
ax3.set_ylabel('Tiempos (s)')
ax3.set_title('Hashing Insert 10000')

# Quince Mil insert
ax4.plot(QMil['Ejecucion'], QMil['tiempo_segundos'])
ax4.set_xlabel('Nmero de intentos')
ax4.set_ylabel('Tiempos (s)')
ax4.set_title('Hashing Insert 15000')

# Veinte Mil insert
ax5.plot(VMil['Ejecucion'], VMil['tiempo_segundos'])
ax5.set_xlabel('Nmero de intentos')
ax5.set_ylabel('Tiempos (s)')
ax5.set_title('Hashing Insert 20000')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```
